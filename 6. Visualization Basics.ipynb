```python
# EDA
# 1. Numerical EDA
# 2. Visual EDA
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.info()
```

```python
tips.head()
```

```python
#Plotting the data
# Numerical Data
# Categorical Data

# If dealing with (Continuous) Numerical Data ----> Use Distribution Plots like Histogram \ KDE
#

sns.distplot(tips['total_bill'])
```

```python
# The above gave me a suspision of outlier. Thus to check the same we shall use
# Numerical EDA form:
# a. describe
# b. 1.5 IQR Rule (Tukey's Rule)
# Visual EDA:
# c. Box and Whiskers Plot

sns.boxplot(x=tips['total_bill'])
```

```python
sns.boxplot(x=tips['tip'])
```

```python
# Apply Box plot on all numerical column and compare outcome with Tukey's Rule
```

```python
# For Numerical data you can also perform Correlation Analysis Visually
# jointplot

sns.jointplot(x = tips['tip'] , y= tips['total_bill'])
```

```python
sns.jointplot(x = tips['tip'] , y= tips['total_bill'], kind='reg')
```

```python
sns.jointplot(x = tips['tip'] , y= tips['total_bill'], kind='hex')
```
